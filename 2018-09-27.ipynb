```python
import numpy as np
import pandas as pd
import glob, os, sys

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
src_dir = "/mnt/recordings/User/tomii/Takanari/2018-09-03/csv"
```

```python
files = sorted(glob.glob(src_dir+"/*.csv"))
for f in files : print(f)
```

```python
df = pd.read_csv(os.path.join(src_dir, "Datasheet.csv"))
df
```

```python
switch = {
    "show_column" : True,
    "plot_all" : False,
}
```

```python
for i, row in df.iterrows():
    df_ = pd.read_csv(os.path.join(src_dir, "No.{0:0>2}.csv".format(i)))
    
    if switch["show_column"]:
        print(i, df_.columns)
    
    dat = df_.values
    
    if switch["plot_all"]:
        plt.clf()
        for s in dat.T:
            plt.plot(s)
        plt.show()
```
